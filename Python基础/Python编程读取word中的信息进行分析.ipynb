## 代码示例：读取段落信息

```python
from docx import Document
import tqdm

# 打开文档1
doc1 = Document('data/首批国家级一流本科课程名单.docx')

# 读取每段内容
pl = [ paragraph.text for paragraph in doc1.paragraphs]

print('###### 输出word1文章内容')
# 输出读取到的内容
i = 0
for t in pl:
    i += 1
    print(f"表{i}:{t}")


```

## 代码示例2 读取表格信息

```python
# 读取表格材料，并输出结果
universities = [] # 大学名
courses = [] # 课程名
authors = [] # 主持人
tables = [table for table in doc1.tables]

'''
以下示例针对一张表进行展示
对全部表格进行统计代码改为：
for table in tables:
'''
for table in tables[4:5]:
    for row in tqdm.tqdm(table.rows): # 使用进度条
        universities.append(row.cells[4].text)
        courses.append(row.cells[1].text)
        authors.append(row.cells[2].text)
        '''for cell in row.cells:
            
            if '陕西师范大学' in cell.text:
                print (cell.text,end='  ')
            
                print()
                print('\n')'''
print(universities)
print(authors)
print(courses)
```
