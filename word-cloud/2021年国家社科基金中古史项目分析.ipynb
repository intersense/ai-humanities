## pandas读取HTML的表格
pandas的read——html()函数是一个快速、方便的将HTML表格转换为Pandas DataFrame的方法。我们使用该工具可以方便地提取网站中的表格，而无需手动解析HTML源码。

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from unicodedata import normalize
import jieba
from wordcloud import WordCloud
from PIL import Image
```

```python
text = ""
for tb in table_MN:
    print(tb[0][0])
    text += str(tb[0][0])
```

```python
sep_list=jieba.lcut_for_search(text,)
sep_list
```

```python
def stopwordslist():
    stopwords = [line.strip() for line in open('stopwords.txt',encoding='UTF-8').readlines()]
    return stopwords



stopwords = stopwordslist()
#stopwords
words = [word for word in sep_list if word not in stopwords and len(word) > 1] # 去除停用词和去除单个文字
print(f"总共词数：{len(words)}")
print(words)
```

```python
words1 = ""
for word in words:
    words1 += " "
    words1 += word
words1
```

```python
#图片获取
mask=np.array(Image.open("background.png"))#--这是词云的背景图形状

# 数据清洗
# 屏蔽45
#STOPWORDS.add('45')

font=r'SourceHanSerifK-Light.otf'#---（必须引用字体不然代码会报错）

wc=WordCloud(
    scale=4,#调整图片大小---（如果设置太小图会很模糊）
    font_path=font,#使用的字体库
    max_words=200,  # 词云显示的最大词数
    margin=2,#字体之间的间距
    mask=mask,#背景图片
    background_color='white', #背景颜色
    max_font_size=200,
    # min_font_size=1,
    # stopwords=STOPWORDS, #屏蔽的内容
    collocations=False, #避免重复单词
    width=1600,height=1200 #图像宽高，字间距
)

wc.generate(words1) #制作词云
wc.to_file('词云.jpg') #保存到当地文件

# 图片展示
plt.figure(dpi=400) #通过这里可以放大或缩小
plt.imshow(wc,interpolation='catrom')
plt.axis('off')
plt.show()
```
