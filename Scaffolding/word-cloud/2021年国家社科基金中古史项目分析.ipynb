## pandas读取HTML的表格
pandas的read_html()函数是一个快速、方便的将HTML表格转换为Pandas DataFrame的方法。我们使用该工具可以方便地提取网站中的表格，而无需手动解析HTML源码。

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import jieba
from wordcloud import WordCloud
from PIL import Image
```

```python
table_MN = pd.read_html('https://mp.weixin.qq.com/s/gyLAm2qV5CHGzFE9D80JzQ')
print(f"Total Tables:{len(table_MN)}")
```

```python
df = table_MN[0]
df.head()
```

```python
text = ""
proj_num = 0
for tb in table_MN:
    
    if len(tb[0]) > 1:
        #print(tb[0])
        for row in tb[0]:
            #print(row[0])
            text += str(row[0])
            proj_num += 1
    if len(tb[0]) == 1:
        #print(tb[0][0])
        text += str(tb[0][0])
        proj_num += 1
print(f"项目数:{proj_num}")
print(text)
```

```python
sep_list=jieba.lcut_for_search(text,)
sep_list
```

```python
def stopwordslist():
    stopwords = [line.strip() for line in open('stopwords.txt',encoding='UTF-8').readlines()]
    return stopwords



stopwords = stopwordslist()
#stopwords
words = [word for word in sep_list if word not in stopwords and len(word) > 1] # 去除停用词和去除单个文字
print(f"总共词数：{len(words)}")
print(words)
```

```python
words1 = ""
for word in words:
    words1 += " "
    words1 += word
words1
```

```python
# 设置词云的背景图形状
mask=np.array(Image.open("background.png"))
# 设置词云文字的字体
font=r'SourceHanSerifK-Light.otf'

wc=WordCloud(
    scale=4, #调整图片大小
    font_path=font, #使用的字体库
    max_words=200,  # 词云显示的最大词数
    margin=2, #字体之间的间距
    mask=mask, #背景形状图片
    background_color='white', #背景颜色
    max_font_size=200,
    collocations=False, #避免重复单词
    width=1600,height=1200 #图像尺寸 宽和高
)

wc.generate(words1) #制作词云
wc.to_file('词云.jpg') #保存到当地文件

# 图片展示
plt.figure(dpi=400) # 参数dpi设置图形每英寸的点数
plt.imshow(wc, interpolation='catrom')
plt.axis('off')
plt.show()
```
